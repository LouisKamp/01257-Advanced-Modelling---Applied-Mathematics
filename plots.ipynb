```python
import numpy as np
import matplotlib.pyplot as plt
import math
from pendulum_mag_sim import Pendulum

phis = np.linspace(-np.pi,np.pi,20)
dphis = np.linspace(-1,1,20)

# Constants
M_p = 1.0
m_p = 1
g = 1.0
l = 1.0
m_p_abs = 1.0
mu_0 = 1.0
m_1_x = 1.0
y_1 = 1.0
x_1 = 1.0
m_1_y = 1.0
alpha = 0.1


def V(phi):
    return -M_p*g*l*math.cos(phi) + (1/4)*math.pi*m_p_abs*mu_0*((m_1_x*((l*math.sin(phi) - y_1)**2 + (l*math.cos(phi) - x_1)**2) - 3*(l*math.cos(phi) - x_1)*(m_1_x*(l*math.cos(phi) - x_1) + m_1_y*(l*math.sin(phi) - y_1)))*math.cos(phi) + (m_1_y*((l*math.sin(phi) - y_1)**2 + (l*math.cos(phi) - x_1)**2) - 3*(l*math.sin(phi) - y_1)*(m_1_x*(l*math.cos(phi) - x_1) + m_1_y*(l*math.sin(phi) - y_1)))*math.sin(phi))/((l*math.sin(phi) - y_1)**2 + (l*math.cos(phi) - x_1)**2)**(5/2)

for phi, dphi in zip(phis, dphis):
    sim = Pendulum(M_p=M_p,m_p=m_p,l=l,mu_0=mu_0,g=g,alpha=0.1)
    phi_res, dphi_res = sim.simulate((0,100), (phi, dphi), show_plots=False)
    Vs = [V(phi) for phi in phi_res]
   # plt.scatter(phi_res[0], Vs[0], color='red', s=10)
   # plt.scatter(phi_res[-1], Vs[-1], color='blue', s=10)
    plt.plot(phi_res, Vs, linewidth=0.5)


plt.xlabel("Phi")
plt.ylabel("Potential Energy")
plt.show()
```
